# 新聞資料清理

```python
import pandas as pd

news_df = pd.read_csv('RedditNews(train).csv', encoding='utf-8')
print(news_df.head())
```

```python
import re

def remove_special_characters(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)

news_df['News']=news_df['News'].apply(remove_special_characters)
print(news_df.head())
```

```python
# 查看每日新聞數量
date_counts = news_df['Date'].value_counts().reset_index()
# 根據日期排序
date_counts.sort_values(by='Date', ascending=True)
```

```python
news_df['Date'] = pd.to_datetime(news_df['Date'])
combined_news = news_df.groupby('Date')['News'].apply(' '.join).reset_index()
combined_news
```

```python
stock_data = pd.read_csv('DJIA_table(train).csv', encoding='utf-8')
# 日期欄位轉換成 datetime 格式
stock_data['Date'] = pd.to_datetime(stock_data['Date'], dayfirst=True)
# 設定日期為索引
stock_data.set_index('Date', inplace=True)
# 根據日期排序
stock_data = stock_data.sort_index()
# 印出前10筆
stock_data.head(10)
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# 如果尚未下載 VADER 詞典，請執行以下命令
nltk.download('vader_lexicon')

# 初始化情緒分析器
sid = SentimentIntensityAnalyzer()

# 定義一個函數來計算情緒得分
def get_sentiment(text):
    return sid.polarity_scores(text)['compound']

# 對合併後的新聞資料進行情緒分析
combined_news['Sentiment'] = combined_news['News'].apply(get_sentiment)

# 查看結果
combined_news.head()
```

```python
stock_data.reset_index(inplace=True)

# 計算 'Close' 價格的百分比變化
stock_data['Price_Change'] = stock_data['Close'].pct_change()

# 將 'Price_Change' 列向上移動一行，獲取下一天的變化
stock_data['Next_Day_Change'] = stock_data['Price_Change'].shift(-1)

# 對於分類任務，創建一個二元目標變量
stock_data['Target'] = stock_data['Next_Day_Change'].apply(lambda x: 1 if x > 0 else 0)

# 查看結果
stock_data.head()
```

```python
merged_data = pd.merge(stock_data, combined_news[['Date', 'Sentiment']], on='Date', how='inner')

# 刪除任何缺失值
merged_data.dropna(inplace=True)

# 查看結果
merged_data.head()
```

```python
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'Sentiment']

# 目標變量
target = 'Target'

# 準備特徵矩陣 X 和目標向量 y
X = merged_data[features]
y = merged_data[target]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# 分割資料集，不打亂順序以保留時間序列特性
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# 初始化模型
model = RandomForestClassifier(n_estimators=100, random_state=42)

# 訓練模型
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# 在測試集上進行預測
y_pred = model.predict(X_test)

# 計算準確率
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# 顯示詳細的分類報告
print(classification_report(y_test, y_pred))
```
