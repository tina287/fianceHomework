```python
import yfinance as yf
from datetime import datetime

def get_tsmc_stock_data():
    # 獲取台積電股票的 ticker 對象
    tsmc = yf.Ticker("2330.TW")

    # 獲取當前股票信息
    info = tsmc.info

    # 獲取最新的股價數據
    history = tsmc.history(period="1d")

    # 獲取當前時間
    current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    return {
        "時間": current_time,
        "股價": history['Close'].iloc[-1] if not history.empty else "N/A",
        "漲跌": f"{history['Close'].iloc[-1] - history['Open'].iloc[-1]:.2f}" if not history.empty else "N/A",
        "成交量": info.get('volume', "N/A"),
        "52週高點": info.get('fiftyTwoWeekHigh', "N/A"),
        "52週低點": info.get('fiftyTwoWeekLow', "N/A"),
        "市值": info.get('marketCap', "N/A")
    }

if __name__ == "__main__":
    data = get_tsmc_stock_data()
    print("台積電(2330.TW)股市資料:")
    for key, value in data.items():
        print(f"{key}: {value}")
```
